```python
messages_df = pd.read_excel("/home/zorinep/big_data_magistracy/vizualize/datasets/messages-2023-with-hours.xlsx")
messages_df.columns = ["date_hour",	"message_count"]
messages_df.head()
```

```python
coffee_and_wine_df = pd.read_excel("./datasets/kofe-vino.xlsx")
coffee_and_wine_df.head()
```

```python
daily_sum = coffee_and_wine_df.groupby('Дата')[['Кофе', 'Вино']].sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_sum.index, daily_sum['Кофе'], marker='o', label='Кофе', color='brown')
plt.plot(daily_sum.index, daily_sum['Вино'], marker='o', label='Вино', color='purple')
plt.title('Динамика потребления кофе и вина по дням')
plt.xlabel('Дата')
plt.ylabel('Суммарное потребление')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
vehicle_df = pd.read_excel("./datasets/passazhirooborot.xlsx", )
vehicle_df.head()
# vihicle_df = vihicle_df.T
# vihicle_df = vihicle_df.iloc[1::]
# vihicle_df.columns = []
```

```python
cols = vehicle_df["Тип транспорта"].unique().tolist()
cols
```

```python
vehicle_df = vehicle_df.T
vehicle_df = vehicle_df.iloc[1:]
vehicle_df.columns = cols
```

```python
cols = [
    'дата',
    'железнодорожный',
    'автобусный',
    'трамвайный',
    'троллейбусный',
    'метрополитен',
    'морской',
    'внутренний водный3)',
    'воздушный'
 ]
```

```python
vehicle_df = vehicle_df.reset_index()
```

```python
vehicle_df.columns
```

```python
vehicle_df
```

```python
vehicle_df.set_index('index', inplace=True)

plt.figure(figsize=(12, 8))

# Рисуем линии для каждого вида транспорта
for column in vehicle_df.columns:
    plt.plot(vehicle_df.index, vehicle_df[column], marker='o', label=column, linewidth=2)

# Настройки графика
plt.title('Объем перевозок пассажиров по видам транспорта (2000-2003 гг.)', fontsize=14)
plt.xlabel('Год')
plt.ylabel('Объем перевозок, млрд. пасс.-км')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.grid(True, alpha=0.3)

plt.tight_layout() 
plt.show()
```
